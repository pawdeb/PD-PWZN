```python
# -*- coding: utf-8 -*-

# PWZN Projekt 11
# Paweł Dębowski

import networkx as nx
import matplotlib.pyplot as plt

# DANE
g = nx.read_edgelist("data.txt", nodetype = int)
nx.info(g)
```

```python
# ROZKŁAD STOPNI WĘZŁÓW, W SENSIE HISTOGRAM
degrees = [g.degree(n) for n in g.nodes()]
plt.figure(figsize=(12, 8))
binwidth = 50 
plt.hist(degrees, bins=range(min(degrees), max(degrees) + binwidth, binwidth))
plt.show()
```

```python
# ROZKŁAD STOPNI, W SENSIE WYKRES
degree_freq = nx.degree_histogram(g)
degrees = range(len(degree_freq))
count = len(degree_freq)
prob = [x / count for x in degree_freq]

plt.figure(figsize=(12, 8)) 
plt.loglog(degrees, prob, '.') 
plt.xlabel('Stopień węzła k')
plt.ylabel('Pstwo wystąpienia P(k)') 
```

```python
# SPÓJNE SKŁADOWE I ICH ROZKŁAD
print([len(c) for c in sorted(nx.connected_components(g), key=len, reverse=True)])
# wypluwa jednen komponent (spójny podgraf) czyli nie ma co badać, sieć spójna
```

```python
# ŚREDNIA NAJKRÓTSZA SCIEŻKA
nx.average_shortest_path_length(g) # daje 3.6925068496963913
# liczy się minutę ale potwierdza "sześć połączeń Miligrana"
```

```python
# WIZUALIZACJA 
plt.figure(3,figsize=(12,8)) 
nx.draw(g,node_size=30)
plt.show()

# W CYTOSCAPE JEST TEŻ
```
