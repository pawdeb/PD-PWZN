```python
# -*- coding: utf-8 -*-

# PWZN Projekt 10
# Paweł Dębowski

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Dane
df = pd.read_csv('owid-co2-data.csv')
df
```

```python
# Kraje, gdzie prowadzono statystyki
df['country'].unique()
```

```python
# Chalupy welcome to
df.loc[df['country'] == 'Poland']
```

```python
# Grupowanie i agregacja, czyli top3 substancji
df_co2 = df.groupby('country').agg(sum_co2 = ('co2', np.sum))#
df_methane = df.groupby('country').agg(sum_methane = ('methane', np.sum))
df_no = df.groupby('country').agg(sum_no = ('nitrous_oxide', np.sum))

df_top3 = pd.concat([df_co2, df_methane, df_no], axis=1, join="inner")
df_top3
```

```python
# Gdzie jest najwięcej krów?
df_clean = df[df['iso_code'].notna()] 
df_clean = df[df['iso_code'] != 'OWID_WRL']
# Ale bez kontynentów czy świata!
df_meth = df_clean.groupby('country').agg(sum_methane = ('methane', np.sum))
df_topMeth = df_meth.sort_values('sum_methane').tail(10)
df_topMeth
```

```python
# Piękny wykres na koniec
kraje = df_topMeth.index.to_numpy()
kraje = np.flip(kraje)

f = plt.figure(figsize=(21,9))
for i in kraje:
    y = df.loc[df['country'] == i]['co2'].values
    x = df.loc[df['country'] == i]['year'].values
    plt.plot(x,y, label = i)
plt.legend(prop={'size': 12})
plt.xlabel("Lata")
plt.ylabel("Produkcja metanu")
plt.show()
```
